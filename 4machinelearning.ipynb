```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objs as go
import plotly.io as pio
from sklearn.linear_model import LinearRegression
import kaggle
```

```python
df = df[df['PartnerISO3'] == 'WLD']
```

## 4. 1 Germany

```python
germany_ml = df[(df['ReporterName'] == 'Germany') & (df['Year'].between(2000, 2020))]
x = germany_ml[['Year']]
y = germany_ml['TradeValue in 1000 USD'].values

plt.scatter(x, y)

file_path = 'graphs/ML/germany_past.png'
plt.savefig(file_path)

plt.show()
```

```python
model = LinearRegression()

model.fit(x, y)

print(f"Intercept: {model.intercept_}")
print(f"Slope: {model.coef_}")
```

```python
future_years = pd.DataFrame({'Year': np.arange(2021, 2031)})

future_predictions_germany = model.predict(future_years)

for year, prediction in zip(future_years['Year'], future_predictions_germany):
    print(f"Prediction for {year}: {prediction:.2f} in 1000 USD")
```

```python
plt.scatter(x, y, color='blue', label='Historical Data')

plt.plot(x, model.predict(x), color='green', label='Regression Line')

plt.scatter(future_years, future_predictions_germany, color='red', label='Future Predictions')

plt.xlabel('Year')
plt.ylabel('TradeValue in 1000 USD')
plt.title('Forecasting Future Exports')
plt.legend()

file_path = 'graphs/ML/germany_future.png'
plt.savefig(file_path)

plt.show()
```

## 4. 4 Comparative

```python
def predict_future_trade(df, country_name):
    x = df[['Year']]
    y = df['TradeValue in 1000 USD'].values
    
    model = LinearRegression()
    model.fit(x, y)
    
    future_years = pd.DataFrame({'Year': np.arange(2021, 2031)})
    future_predictions = model.predict(future_years)
    
    predictions_df = pd.DataFrame({
        'Year': future_years['Year'],
        'Predicted TradeValue (in 1000 USD)': future_predictions
    })
    
    return predictions_df

germany_predictions = predict_future_trade(germany_ml, 'Germany')
china_predictions = predict_future_trade(china_ml, 'China')
usa_predictions = predict_future_trade(usa_ml, 'USA')
```

```python
germany_trace = go.Scatter(
    x=germany_predictions['Year'],
    y=germany_predictions['Predicted TradeValue (in 1000 USD)'],
    mode='lines+markers',
    name='Germany'
)

china_trace = go.Scatter(
    x=china_predictions['Year'],
    y=china_predictions['Predicted TradeValue (in 1000 USD)'],
    mode='lines+markers',
    name='China'
)

usa_trace = go.Scatter(
    x=usa_predictions['Year'],
    y=usa_predictions['Predicted TradeValue (in 1000 USD)'],
    mode='lines+markers',
    name='USA'
)

fig = go.Figure(data=[germany_trace, china_trace, usa_trace])

fig.update_layout(
    title='Forecast of Future Exports (Germany, China, USA)',
    xaxis_title='Year',
    yaxis_title='TradeValue in 1000 USD',
    legend_title='Country',
    template='plotly'
)

fig.show()

pio.write_html(fig, file=f'graphs/ML/comparative.html', auto_open=False)
```
